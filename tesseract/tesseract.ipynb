```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from pytesseract.pytesseract import image_to_data, Output
```

```python
# read the image
img = cv.imread("test_image.png", cv.IMREAD_UNCHANGED)
```

```python
def preprocess(img):
    grayed = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    kernel = np.ones((2, 2), dtype=np.uint8)
    eroded = cv.erode(grayed, kernel)
    _, threshed = cv.threshold(grayed, 200, 255, cv.THRESH_BINARY)

    return threshed
```

```python
img_processed = preprocess(img)
plt.imshow(img_processed)
```

```python
ocr_ = image_to_data(img_processed, lang="en", output_type=Output.DICT)
```
